```python
import numpy as np
import pandas as pd

import pymc3 as pm
import theano.tensor as T
from theano.compile.ops import as_op
import theano

from IPython.display import Image
#import utils
%matplotlib inline


BoundedNormal1 = pm.Bound(pm.Normal, lower=0, upper=1)
BoundedNormal2 = pm.Bound(pm.Normal, lower=5, upper=100)
BoundedNormal3 = pm.Bound(pm.Normal, lower=0, upper=600)

with pm.Model() as model:
    carType = pm.Categorical('carType', p=[.33, 0.33, 0.34], observed=1)
    reliability = pm.Categorical('reliability', p=[0.33, 0.33, 0.34], observed=0)
    maintainabilityCont = BoundedNormal1('maintainabilityCont', mu=(2/3*(2*carType+1)/6 + 1/3*(2*reliability+1)/6), sd=0.1)
    maintainability = pm.Deterministic(
        'maintainability',
        T.switch(
            T.lt(maintainabilityCont, 0.33),
            0,
            T.switch(
                T.lt(maintainabilityCont, 0.66),
                1,
                2
            )
        )
    )
#     milesPerGallon = pm.Deterministic(
#         'milesPerGallon',
#         T.switch(
#             T.eq(carType, 0),
#             BoundedNormal2('a1', mu=35, sd=7.07),
#             T.switch(
#                 T.eq(carType, 1),
#                 BoundedNormal2('a2', mu=28, sd=7.07),
#                 BoundedNormal2('a3', mu=18, sd=7.07)
#             )
#         )
#     )
    annualMaintainanceCost = pm.Deterministic(
        'annualMaintainanceCost',
        T.switch(
            T.eq(maintainability, 0),
            BoundedNormal3('b1', mu=100, sd=10),
            T.switch(
                T.eq(maintainability, 1),
                BoundedNormal3('b2', mu=200, sd=12),
                BoundedNormal3('b3', mu=500, sd=12)
            )
        )
    )

with model:
    SAMPLE_NUM = 10000
    trace = pm.sample(SAMPLE_NUM)
    pm.summary(trace, varnames=['maintainability'], start=1000)
    pm.traceplot(trace[1000:], varnames=['maintainability'])
    pm.traceplot(trace[1000:], varnames=['annualMaintainanceCost'])
    # print(sum(trace['C3_0']) / float(SAMPLE_NUM))
```
